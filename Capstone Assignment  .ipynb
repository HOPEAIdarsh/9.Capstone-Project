# Handling Outliers:


```python
# visualization of the outliers for the numeric : avg_price_per_room, lead_time
import matplotlib.pyplot as plt
plt.boxplot(dataset.avg_price_per_room, showmeans=True)
plt.title('Boxplot of avg_price_per_room')
plt.show()
```

```python
# removing the outliers,Calculate IQR for the column
Q1 = dataset['avg_price_per_room'].quantile(0.25)
Q3 = dataset['avg_price_per_room'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Identify and remove outliers
dataset = dataset[~((dataset['avg_price_per_room'] < (Q1 - 1.5 * IQR)) | (dataset['avg_price_per_room'] > (Q3 + 1.5 * IQR)))]
print(dataset.shape, "2") # (34579, 19)
plt.boxplot(dataset.avg_price_per_room, showmeans=True)
plt.title('Boxplot of avg_price_per_room (After Removing Outliers)')
plt.show()
```

```python
plt.boxplot(dataset.lead_time, showmeans=True)
plt.title('Boxplot of lead_time')
plt.show()
Q1 = dataset['lead_time'].quantile(0.25)
Q3 = dataset['lead_time'].quantile(0.75)
IQR = Q3 - Q1
```

```python
dataset = dataset[~((dataset['lead_time'] < (Q1 - 1.5 * IQR)) | (dataset['lead_time'] > (Q3 + 1.5 * IQR)))]
print(dataset.shape, "3") # (33462, 19)
plt.boxplot(dataset.lead_time, showmeans=True)
plt.title('Boxplot of lead_time (After Removing Outliers)')
plt.show()
```

```python
# Assuming no_of_adults, no_of_children and avg_price_per_room should be >= 0
dataset=  dataset[ dataset['no_of_adults'] >= 0]
dataset =  dataset[ dataset['no_of_children'] >= 0]
dataset =  dataset[ dataset['avg_price_per_room'] > 0]
dataset.shape, "4" # (33462, 19)

dataset =  dataset.drop_duplicates()
dataset.shape, "5" # (33462, 19)
```

```python
# converting categorical columns to numeric , dropping the id column

dataset = dataset.drop('Booking_ID', axis=1)
```

```python
# checking the relation between lean time and cancel
bins = [0, 7, 30, 90, float('inf')]
labels = ['Week', 'Month', '3 Months', 'More than 3 months']
```

```python
# Create a new column for lead_time bins
dataset['lead_time_group'] = pd.cut(dataset['lead_time'], bins=bins, labels=labels, right=False)
```

```python
# Initialize data for the stacked bar chart
bar_data = {}
for group in labels:
    subset = dataset[dataset['lead_time_group'] == group]
    bar_data[group] = {
        'Not_Canceled': len(subset[subset['booking_status'] == 'Not_Canceled']),
        'Canceled': len(subset[subset['booking_status'] == 'Canceled'])
    }

# Data to plot
labels = list(bar_data.keys())
not_canceled_counts = [bar_data[label]['Not_Canceled'] for label in labels]
canceled_counts = [bar_data[label]['Canceled'] for label in labels]
```

```python
# Create the stacked bar plot
fig, ax = plt.subplots()
width = 0.5

p1 = plt.bar(labels, not_canceled_counts, width, color='green', label='Not_Canceled')
p2 = plt.bar(labels, canceled_counts, width, color='red', bottom=not_canceled_counts, label='Canceled')

# Add title and labels
plt.title('Booking Status by Lead Time Groups', fontdict={'fontsize': 16})
plt.xlabel('Lead Time Group', fontdict={'fontsize': 14})
plt.ylabel('Number of Bookings', fontdict={'fontsize': 14})

# Add a legend
plt.legend(title='Booking Status')

# Show the plot
plt.show()


dataset['market_segment_type'].value_counts()
```

# Cancellation

```python
# Initialize data for the bar chart
bar_data = {}
market_segments = dataset['market_segment_type'].unique()

for segment in market_segments:
    subset = dataset[dataset['market_segment_type'] == segment]
    total_bookings = len(subset)
    canceled_bookings = len(subset[subset['booking_status'] == 'Canceled'])

    if total_bookings > 0:  # Avoid division by zero
        cancel_percentage = (canceled_bookings / total_bookings) * 100
    else:
        cancel_percentage = 0

    bar_data[segment] = cancel_percentage
# Data to plot
labels = list(bar_data.keys())
cancel_percentages = [bar_data[label] for label in labels]

# Create the bar plot
fig, ax = plt.subplots()
plt.bar(labels, cancel_percentages, color='blue')  # You can change the color as needed

# Add title and labels
plt.title('Percentage of Cancellations ', fontdict={'fontsize': 16})
plt.xlabel('Market Segment', fontdict={'fontsize': 14})
plt.ylabel('Percentage of Cancellations', fontdict={'fontsize': 14})

# Annotate each bar with the corresponding percentage value
for i, v in enumerate(cancel_percentages):
    ax.text(i, v+0.2, "{:.2f}%".format(v), color='black', ha='center')
# Show the plot
plt.show()
```
